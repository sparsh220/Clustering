 # **Theoretical Questions**

## 1. What is unsupervised learning in the context of machine learning ?

Answer : Unsupervised learning is a type of machine learning where the model learns without labeled data — meaning, we don’t give it answers during training.

The model looks for patterns or structures in the data on its own.



## 2. How does K-Means clustering algorithm work?


Answer : K-Means is an unsupervised learning algorithm used for clustering — it groups similar data points into K clusters.

K-Means clustering algorithm work as

* Choose K — the number of clusters you want.

* Pick K random points as the first “centroids” (centers of clusters).

* Assign each data point to the nearest centroid (based on distance, usually Euclidean).

* Recalculate centroids — find the average position of all points in each cluster.

* Repeat steps 3 and 4 until the centroids stop changing (or change very little).

Finally, you get K groups where each group has points that are similar to each other.

## 3. Explain the concept of a dendrogram in hierarchical clustering ?


Answer : A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are grouped together step by step.

## 4. What is the main difference between K-Means and Hierarchical Clustering ?

Answer :  
* K-Means:

    * You must choose the number of clusters (K) before starting.

    * It moves points and centroids around until clusters are stable.

    * It’s fast and works well for large datasets.

*  Hierarchical Clustering:

    * You don’t need to choose K at the start.

    * It builds a tree (dendrogram) by merging or splitting clusters step by step.

    * It’s slower, but it shows the relationship between clusters clearly.

## 5. What are the advantages of DBSCAN over K-Means ?

Answer : The advantages of DBSCAN over K-Means :
1. No need to choose K:
    * You don’t have to say how many clusters you want — DBSCAN finds them automatically.

2.  Can find any shape of cluster:
    * K-Means works best with round clusters, but DBSCAN can handle irregular or curved shapes.

3. Handles noise/outliers better:
    * It can mark outliers as “noise” instead of forcing them into a cluster.


## 6. When would you use Silhouette Score in clustering ?

Answer : We use the Silhouette Score to measure how good the clusters are — it checks how well each data point fits within its cluster compared to other clusters.



## 7. What are the limitations of Hierarchical Clustering ?


Answer :  The limitations of Hierarchical Clustering are :

* Not good for large datasets:
    * It’s slow and memory-heavy because it compares every pair of points.

* No way to fix mistakes:
    * Once two clusters are merged, you can’t undo it — even if it was a bad merge.

* Sensitive to noise and outliers:
    * A few odd points can change the cluster structure a lot.




## 8. Why is feature scaling important in clustering algorithms like K-Means ?

Answer : Feature scaling is very important in clustering algorithms like K-Means because these algorithms depend on distance measurements (mostly Euclidean distance) to decide how similar or different data points are.



## 9. How does DBSCAN identify noise points ?


Answer : DBSCAN identifies noise points (also called outliers) based on how many neighbors a point has within a certain distance.


## 10. Define inertia in the context of K-Means ?

Answer :  In K-Means clustering, inertia is a measure of how well the data points are grouped within their respective clusters. It represents the sum of the squared distances between each data point and the centroid (center) of the cluster it belongs to.

Mathematically, it shows how compact the clusters are — that is, how close the points are to their cluster centers.



## 11. What is the elbow method in K-Means clustering ?


Answer : The elbow method is one of the most popular techniques used to find the optimal number of clusters in K-Means clustering.

In K-Means, we must choose how many clusters (k) to divide the data into. If we choose too few clusters, different groups may get mixed together; if we choose too many, the model becomes too complex. The elbow method helps to balance this.



## 12. Describe the concept of "density" in DBSCAN ?


Answer : In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), the idea of density is very important. It helps the algorithm decide which points belong to the same cluster.


Density means how many data points are close to each other within a certain distance.
If many points are packed close together, that area is called a dense region, and DBSCAN treats it as a cluster.

In short DBSCAN forms clusters based on areas of high density and separates them from areas of low density (noise).

## 13. Can hierarchical clustering be used on categorical data ?

Answer : Yes, hierarchical clustering can be used on categorical data, but not directly in the same way as with numeric data

## 14. What does a negative Silhouette Score indicate ?


Answer :  A negative Silhouette Score indicates that the clustering result is poor, and many data points may have been assigned to the wrong clusters.

## 15. Explain the term "linkage criteria" in hierarchical clustering ?


Answer : In hierarchical clustering, the linkage criteria decide how the distance between two clusters is calculated when forming the hierarchy.

When you merge clusters step by step, you need to know which two clusters are closest. The linkage method tells you how to measure that closeness.

## 16. Why might K-Means clustering perform poorly on data with varying  cluster sizes or densities ?


Answer : K-Means clustering may perform poorly on datasets that contain clusters of different sizes or densities because of the basic assumptions and working principle of the algorithm.

K-Means works by dividing data into k clusters such that each point belongs to the nearest centroid (mean point of a cluster). It assumes that all clusters are spherical, equally sized, and have similar densities. This assumption causes problems when the actual data does not follow this pattern.


## 17. What are the core parameters in DBSCAN, and how do they influence clustering ?


Answer :  In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), there are two main core parameters are :

* Epsilon:

     * It defines the radius (neighborhood) around a data point.

    * A small epsilon value means the algorithm will find many small clusters or even treat most points as noise.

    * A large epsilon value means clusters will be larger and may merge together.

* MinPts (Minimum Points):

     * It is the minimum number of points required within the e-radius to form a dense region (a cluster).

    * If a point has at least MinPts neighbors inside its e area, it becomes a core point.

     * A low MinPts value can lead to detecting too many small clusters (even noise as clusters).

     * A high MinPts value can make it harder to form clusters, marking many points as noise.

They influence clustering  as :

* A point is a core point if it has at least MinPts neighbors within e.

* A point close to a core point but with fewer neighbors is a border point.

* A point that doesn’t satisfy either condition is treated as noise or outlier.

## 18.  How does K-Means++ improve upon standard K-Means initialization ?


Answer :  K-Means++ is an improved version of the standard K-Means algorithm that focuses on making the initialization of centroids smarter and more efficient.

In standard K-Means, the initial centroids (cluster centers) are chosen randomly from the data points. This random choice can often lead to poor clustering.

K-Means++ improves standard K-Means by choosing better initial centroids that are far apart and representative, which results in faster, more reliable, and higher-quality clustering outcomes

## 19.  What is agglomerative clustering ?


Answer : Agglomerative clustering is a popular type of hierarchical clustering method that follows a bottom-up approach to form clusters. It starts by treating each data point as a separate cluster and then gradually merges the clusters based on their similarity until all points belong to one single cluster or until a desired number of clusters is reached.

## 20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer : The Silhouette Score is considered a better metric than inertia for evaluating clustering quality because it gives more meaningful information about how well the data points fit within their clusters and how separate the clusters are.


While inertia measures only how compact the clusters are, the Silhouette Score measures both compactness and separation, giving a more complete and accurate picture of how well the data has been clustered. Hence, it is a better and more balanced metric for evaluating clustering performance.

```python
# 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot


from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=400, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
y = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X')
plt.show()
```

```python
# 29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Create data
X, _ = make_blobs(n_samples=400, centers=4, random_state=42)

# Evaluate silhouette scores for k = 2 to 5
scores = []
for k in range(2, 6):
    y = KMeans(n_clusters=k, random_state=42).fit_predict(X)
    scores.append(silhouette_score(X, y))

# Plot bar chart
plt.bar(range(2, 6), scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()


```

```python
#30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Load data
X = load_iris().data

# Hierarchical clustering (average linkage)
Z = linkage(X, method='average')

# Plot dendrogram
plt.figure(figsize=(8,4))
dendrogram(Z)
plt.title("Hierarchical Clustering (Average Linkage)")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Create overlapping clusters
X, _ = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

# Plot decision boundaries
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c='red', s=200, marker='X')
plt.title("K-Means with Decision Boundaries")
plt.show()

```
